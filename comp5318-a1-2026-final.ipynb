## **1. Data Pre-processing**

```python
# Import all libraries
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
```

```python
# Ignore future warnings
from warnings import simplefilter
simplefilter(action='ignore', category=FutureWarning)
```

```python
# Load the rice dataset: rice-final2.csv
dataset = pd.read_csv("rice-final2.csv", na_values="?")

feature_columns = dataset.columns[:-1]
target_column = dataset.columns[-1]

X_raw = dataset[feature_columns]
y_raw = dataset[target_column]
```

```python
# Pre-process dataset
imputer = SimpleImputer(strategy="mean")
scaler = MinMaxScaler()

X = imputer.fit_transform(X_raw)
X = scaler.fit_transform(X)

class_mapping = {"class1": 0, "class2": 1}
y = y_raw.replace(class_mapping).astype(int).to_numpy()
```

```python
# Print first ten rows of pre-processed dataset to 4 decimal places as per assignment spec
# A function is provided to assist

def print_data(X, y, n_rows=10):
    """Takes a numpy data array and target and prints the first ten rows.
    
    Arguments:
        X: numpy array of shape (n_examples, n_features)
        y: numpy array of shape (n_examples)
        n_rows: numpy of rows to print
    """
    for example_num in range(min(n_rows, len(X))):
        for feature in X[example_num]:
            print("{:.4f}".format(feature), end=",")

        if example_num == len(X)-1:
            print(y[example_num],end="")
        else:
            print(y[example_num])

print_data(X, y)
```

## **2. Build Classifiers**

- Part 1:  Logistic Regression, Naïve Bayes
- Part 2:  KNN, Decision Tree, Ada Boost, Gradient Boost, Random Forest, SVM

### Part 1: Cross-validation without parameter tuning

```python
## Setting the 10 fold stratified cross-validation
cvKFold=StratifiedKFold(n_splits=10, shuffle=True, random_state=0)

# The stratified folds from cvKFold should be provided to the classifiers
```

```python
# Logistic Regression
logr_clf = LogisticRegression(random_state=0, solver="liblinear")
logr_scores = cross_val_score(logr_clf, X, y, cv=cvKFold, scoring="accuracy")
logr_avg_accuracy = np.mean(logr_scores)
```

```python
# Naïve Bayes
nb_clf = GaussianNB()
nb_scores = cross_val_score(nb_clf, X, y, cv=cvKFold, scoring="accuracy")
nb_avg_accuracy = np.mean(nb_scores)
```

### Part 1 Results


```python
# Print results for each classifier in part 1 to 4 decimal places here:
print(f"LogR average cross-validation accuracy: {logr_avg_accuracy:.4f}")
print(f"NB average cross-validation accuracy: {nb_avg_accuracy:.4f}")
```

### Part 2: Cross-validation with parameter tuning

```python
# KNN 
# parameters you may consider
k = [1, 3, 5, 7]
p = [1, 2]

knn_clf = KNeighborsClassifier()
knn_param_grid = {"n_neighbors": k, "p": p}
```

```python
# Decision Tree 
# parameters you may consider
max_depth = [3, 5, 7, 10]
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]

dt_clf = DecisionTreeClassifier(random_state=0)
dt_param_grid = {
    "max_depth": max_depth,
    "min_samples_split": min_samples_split,
    "min_samples_leaf": min_samples_leaf
}
```

```python
# Ada Boost
# parameters you may consider
n_estimators = [50, 100, 150]
learning_rate = [0.1, 0.2, 0.3, 0.5]

ada_clf = AdaBoostClassifier(random_state=0)
ada_param_grid = {"n_estimators": n_estimators, "learning_rate": learning_rate}
```

```python
# Gradient Boost
# parameters you may consider
max_depth = [1, 3, 5, 7]
n_estimators = [50, 100, 150]
learning_rate = [0.1, 0.2, 0.3, 0.5]

gb_clf = GradientBoostingClassifier(random_state=0)
gb_param_grid = {
    "max_depth": max_depth,
    "n_estimators": n_estimators,
    "learning_rate": learning_rate
}
```

```python
# Random Forest
# You should use RandomForestClassifier from sklearn.ensemble with information gain and max_features set to ‘sqrt’.
# parameters you may consider
n_estimators = [10, 30, 60, 100]
max_leaf_nodes = [6, 12]

rf_clf = RandomForestClassifier(
    criterion="entropy", max_features="sqrt", random_state=0
)
rf_param_grid = {"n_estimators": n_estimators, "max_leaf_nodes": max_leaf_nodes}
```

```python
# SVM
# parameters you may consider
C = [0.01, 0.1, 1, 5]
gamma = [0.01, 0.1, 1, 10]
# optional
kernel = ["rbf"]

svm_clf = SVC(random_state=0)
svm_param_grid = {"C": C, "gamma": gamma, "kernel": kernel}
```

### Part 2: Results

```python
# Perform Grid Search with 10-fold stratified cross-validation (GridSearchCV in sklearn). 
# The stratified folds from cvKFold should be provided to GridSearchCV

# This should include using train_test_split from sklearn.model_selection with stratification and random_state=0
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=0
)
```

```python
# KNN
knn_grid = GridSearchCV(knn_clf, knn_param_grid, cv=cvKFold, scoring="accuracy")
knn_grid.fit(X_train, y_train)
knn_test_pred = knn_grid.best_estimator_.predict(X_test)
knn_test_acc = accuracy_score(y_test, knn_test_pred)

print("KNN best k:", knn_grid.best_params_["n_neighbors"])
print("KNN best p:", knn_grid.best_params_["p"])
print(f"KNN cross-validation accuracy: {knn_grid.best_score_:.4f}")
print(f"KNN test set accuracy: {knn_test_acc:.4f}")
```

```python
# Decision Tree
dt_grid = GridSearchCV(dt_clf, dt_param_grid, cv=cvKFold, scoring="accuracy")
dt_grid.fit(X_train, y_train)
dt_test_pred = dt_grid.best_estimator_.predict(X_test)
dt_test_acc = accuracy_score(y_test, dt_test_pred)

print("DT best max_depth:", dt_grid.best_params_["max_depth"])
print("DT best min_samples_split:", dt_grid.best_params_["min_samples_split"])
print("DT best min_samples_leaf:", dt_grid.best_params_["min_samples_leaf"])
print(f"DT cross-validation accuracy: {dt_grid.best_score_:.4f}")
print(f"DT test set accuracy: {dt_test_acc:.4f}")
```

```python
# Ada Boost
ada_grid = GridSearchCV(ada_clf, ada_param_grid, cv=cvKFold, scoring="accuracy")
ada_grid.fit(X_train, y_train)
ada_test_pred = ada_grid.best_estimator_.predict(X_test)
ada_test_acc = accuracy_score(y_test, ada_test_pred)

print("Ada best n_estimators:", ada_grid.best_params_["n_estimators"])
print(f"Ada best learning_rate: {ada_grid.best_params_['learning_rate']:.4f}")
print(f"Ada cross-validation accuracy: {ada_grid.best_score_:.4f}")
print(f"Ada test set accuracy: {ada_test_acc:.4f}")
```

```python
# Gradient Boost
gb_grid = GridSearchCV(gb_clf, gb_param_grid, cv=cvKFold, scoring="accuracy")
gb_grid.fit(X_train, y_train)
gb_test_pred = gb_grid.best_estimator_.predict(X_test)
gb_test_acc = accuracy_score(y_test, gb_test_pred)

print("GB best max_depth:", gb_grid.best_params_["max_depth"])
print("GB best n_estimators:", gb_grid.best_params_["n_estimators"])
print(f"GB best learning_rate: {gb_grid.best_params_['learning_rate']:.4f}")
print(f"GB cross-validation accuracy: {gb_grid.best_score_:.4f}")
print(f"GB test set accuracy: {gb_test_acc:.4f}")
```

```python
# Random Forest
rf_grid = GridSearchCV(rf_clf, rf_param_grid, cv=cvKFold, scoring="accuracy")
rf_grid.fit(X_train, y_train)
rf_test_pred = rf_grid.best_estimator_.predict(X_test)
rf_test_acc = accuracy_score(y_test, rf_test_pred)
rf_macro_f1 = f1_score(y_test, rf_test_pred, average="macro")
rf_weighted_f1 = f1_score(y_test, rf_test_pred, average="weighted")

print("RF best n_estimators:", rf_grid.best_params_["n_estimators"])
print("RF best max_leaf_nodes:", rf_grid.best_params_["max_leaf_nodes"])
print(f"RF cross-validation accuracy: {rf_grid.best_score_:.4f}")
print(f"RF test set accuracy: {rf_test_acc:.4f}")
print(f"RF test set macro average F1: {rf_macro_f1:.4f}")
print(f"RF test set weighted average F1: {rf_weighted_f1:.4f}")
```

```python
# SVM
svm_grid = GridSearchCV(svm_clf, svm_param_grid, cv=cvKFold, scoring="accuracy")
svm_grid.fit(X_train, y_train)
svm_test_pred = svm_grid.best_estimator_.predict(X_test)
svm_test_acc = accuracy_score(y_test, svm_test_pred)

print(f"SVM best C: {svm_grid.best_params_['C']:.4f}")
print(f"SVM best gamma: {svm_grid.best_params_['gamma']:.4f}")
print("SVM best kernel:", svm_grid.best_params_["kernel"])
print(f"SVM cross-validation accuracy: {svm_grid.best_score_:.4f}")
print(f"SVM test set accuracy: {svm_test_acc:.4f}")
```

## **3. Reflection and Discussion**


The assignment compared the performance of various classification models on the same dataset, including Logistic Regression, Naive Bayes, KNN, Decision Tree, AdaBoost, Gradient Boosting, Random Forest, and SVM. A 10-fold stratified cross-validation together with test set evaluation was used to examine differences in performance and generalization ability across the models.

In Part 1, Logistic Regression (0.9386) slightly outperformed Naive Bayes (0.9264), indicating that Logistic Regression serves as a strong and stable baseline model for this dataset. This may be because Logistic Regression can effectively model linear decision boundaries while remaining robust to noise, whereas Naive Bayes relies on the strong independence assumption between features, which may not fully hold for this dataset.

In Part 2, Gradient Boosting achieved the strongest overall performance with the highest cross-validation accuracy (0.9446) and the highest test accuracy (0.9429). This superior performance can be attributed to its ability to iteratively correct errors made by previous models and capture complex non-linear relationships in the data, leading to improved predictive accuracy and generalization. Random Forest also performed well, with high macro F1 (0.9414) and weighted F1 (0.9427), indicating strong stability and good class balance ability. This is because Random Forest reduces variance through bagging and random feature selection, making it less sensitive to overfitting and noise.

Although Decision Tree showed relatively lower cross-validation accuracy, its test accuracy remained high (0.9429). This suggests that the dataset may have relatively simple decision boundaries that can be captured even by a single tree. In addition, the small difference between training and test performance indicates that the model does not suffer from significant overfitting in this case. AdaBoost also performed well and showed stable learning behaviour, as it focuses on misclassified samples during training. However, its performance was slightly inferior to Gradient Boosting, likely because Gradient Boosting optimizes the model more effectively by directly minimizing the loss function.

KNN and SVM performed relatively worse compared to the other models. Although KNN achieved its optimal parameter k = 7, its overall performance remained lower, possibly due to the reduced effectiveness of distance-based methods in higher-dimensional spaces, where distances become less discriminative. SVM achieved a high cross-validation accuracy (0.9429), but its test accuracy was lower (0.9321), indicating a limitation in generalization performance. This may be due to sensitivity to parameter settings and kernel choice, which can affect its ability to generalize well on unseen data.

Overall, ensemble learning methods performed best. Gradient Boosting showed the strongest overall performance due to its ability to capture complex patterns and reduce bias, while Random Forest provided strong and stable performance by reducing variance. Logistic Regression served as a reliable baseline model due to its simplicity and robustness. In contrast, Naive Bayes and KNN were relatively weaker, while SVM showed strong validation performance but slightly lower generalization ability.

Hyperparameter tuning helped the models find better parameter configurations and contributed to more stable model performance. Rather than producing large improvements in accuracy, tuning primarily improved consistency between cross-validation and test results, indicating better generalization. This suggests that the dataset is relatively well-behaved and can already be modeled effectively by baseline configurations, while tuning helps refine model behaviour and reduce performance variance.

Furthermore, the small differences between cross-validation accuracy and test accuracy across models indicate that the models have good generalization ability and that no significant overfitting was observed. This consistency suggests that the models are neither too complex nor underfitting the data, and that the evaluation results are reliable.
